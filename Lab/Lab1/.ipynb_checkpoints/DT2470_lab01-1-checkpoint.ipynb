Below I show the first 0.6 seconds of my audio file, which I created using https://elevenlabs.io "Text to SFX" tool with the prompt: "car going beep beep".

```python
import pydub
import matplotlib.pyplot as plt
import numpy as np

# The following makes the plot look nice
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (10, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)

# add your code below
# audio source: https://freesound.org/people/frenkfurth/sounds/650676/
# read audio 
wav_audio = pydub.AudioSegment.from_file("fart.wav", format="wav")
bit_depth = wav_audio.sample_width * 8
# print(wav_audio.frame_rate)


# sample the audio
samples = np.array(wav_audio.get_array_of_samples())
samples = samples / (2**(bit_depth-1)) # normalize sample

# generate time axis
time_axis = np.linspace(0, len(wav_audio) / 1000, len(samples))
# len(wav_audio) is the length of audio file in milliseconds
# len(samples) is the number of samples

# plot the figure
plt.plot(time_axis, samples)
plt.title('Fart')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

3. For each of the segments you looked at above, window them, and compute their Fourier transforms (hint: see np.fft.fft). Plot their dB magnitude spectra. Appropriately label your axes with "Magnitude (dB)" and "Frequency (kHz)". The frequency axis **must be** in kiloHertz, and limited to 0 to the Nyquist frequency (half the sampling rate). The magnitude axis **must be** in dB. Use the 1) boxcar window and 2) Hann window (hint: see np.hanning).

```python
# add your code below

# zoom from 1(s) to 1.1(s)
# Segment the audio
segment_duration = 0.1  # Duration of the segment in s
segment_start = 1
start_sample = int(sample_rate * segment_start)
end_sample = int(sample_rate * (segment_start + segment_duration))
segment = samples[start_sample:end_sample]

# generate the frequency axis in kHz
freqs = np.fft.fftfreq(len(segment), d=1/sample_rate) / 1000  # Convert to kHz
nyquist_limit = sample_rate / 2000  # in kHz (half the sample rate)

# apply hann window
hann_window = np.hanning(len(segment))
hann_windowed = segment * hann_window

# fourier transform
fft_hann = np.fft.fft(hann_windowed)

# compute Magnitude in dB (only positive half of the spectrum)
magnitude_hann = 20 * np.log10(np.abs(fft_hann[:len(segment) // 2]))

# limit the frequency axis to the positive half (Nyquist limit)
freqs = freqs[:len(segment) // 2]

# plot the magnitude spectrum for the Hann window
plt.plot(freqs, magnitude_hann)
plt.xlim([0, nyquist_limit])
plt.title("Magnitude Spectrum (1s to 1.1s)")
plt.xlabel("Frequency (kHz)")
plt.ylabel("Magnitude (dB)")
plt.grid(True)
plt.show()


# zoom from 2(s) to 2.1(s)
# Segment the audio
segment_duration = 0.1  # Duration of the segment in s
segment_start = 2
start_sample = int(sample_rate * segment_start)
end_sample = int(sample_rate * (segment_start + segment_duration))
segment = samples[start_sample:end_sample]

# generate the frequency axis in kHz
freqs = np.fft.fftfreq(len(segment), d=1/sample_rate) / 1000  # Convert to kHz
nyquist_limit = sample_rate / 2000  # in kHz (half the sample rate)

# apply hann window
hann_window = np.hanning(len(segment))
hann_windowed = segment * hann_window

# fourier transform
fft_hann = np.fft.fft(hann_windowed)

# compute Magnitude in dB (only positive half of the spectrum)
magnitude_hann = 20 * np.log10(np.abs(fft_hann[:len(segment) // 2]))

# limit the frequency axis to the positive half (Nyquist limit)
freqs = freqs[:len(segment) // 2]

# plot the magnitude spectrum for the Hann window
plt.plot(freqs, magnitude_hann)
plt.xlim([0, nyquist_limit])
plt.title("Magnitude Spectrum (2s to 2.1s)")
plt.xlabel("Frequency (kHz)")
plt.ylabel("Magnitude (dB)")
plt.grid(True)
plt.show()

```

4. For the first 10 seconds of your audio file (or shorter if it is not so long), compute and plot its dB magnitude short-time Fourier transform using a Hann window of duration 25 ms with a window hopsize of 10 ms, and an FFT size of 8192 samples (hint: see signal.stft). Do the same using a Hann window of duration 100 ms with a window hopsize of 10 ms. Appropriately label your axes with "Frequency (kHz)" and "Time (s)". The frequency axis must be in kiloHertz, and limited to 0 to 5 kHz. The time axis must be in seconds. Choose a colormap that you feel describes your personality (see https://matplotlib.org/3.1.1/tutorials/colors/colormaps.html). 

```python
# add your code below
# since this audio is less than 10 seconds, we will use the full file for further processing

from scipy import signal

# parameters for STFT
fft_size = 8192
hop_size = int(0.01 * sample_rate)  # 10 ms hopsize

# Hann window of 25 ms
hann_25ms = signal.windows.hann(int(0.025 * sample_rate))  # 25 ms window

# Hann window of 100 ms
hann_100ms = signal.windows.hann(int(0.1 * sample_rate))   # 100 ms window

# STFT with 25 ms Hann window
f_25ms, t_25ms, Zxx_25ms = signal.stft(samples, fs=sample_rate, window=hann_25ms, nperseg=len(hann_25ms), noverlap=len(hann_25ms) - hop_size, nfft=fft_size)

# STFT with 100 ms Hann window
f_100ms, t_100ms, Zxx_100ms = signal.stft(samples, fs=sample_rate, window=hann_100ms, nperseg=len(hann_100ms), noverlap=len(hann_100ms) - hop_size, nfft=fft_size)

# Convert the magnitudes to dB
magnitude_25ms_dB = 20 * np.log10(np.abs(Zxx_25ms) + 1e-6)  # Adding a small constant to avoid log(0)
magnitude_100ms_dB = 20 * np.log10(np.abs(Zxx_100ms) + 1e-6)

# limit freq axis to 0–5 kHz
f_limit = 5000  # 5 kHz
f_25ms_kHz = f_25ms / 1000  # Convert to kHz
f_100ms_kHz = f_100ms / 1000

# Find the index for 5 kHz cutoff
f_index_limit_25ms = np.where(f_25ms_kHz <= f_limit / 1000)[0][-1]
f_index_limit_100ms = np.where(f_100ms_kHz <= f_limit / 1000)[0][-1]


# Plot for 25 ms window
plt.figure(figsize=(10, 6))
plt.pcolormesh(t_25ms, f_25ms_kHz[:f_index_limit_25ms], magnitude_25ms_dB[:f_index_limit_25ms, :], shading='gouraud', cmap='plasma')
plt.title("STFT Magnitude (25 ms)")
plt.ylabel("Frequency (kHz)")
plt.xlabel("Time (s)")
plt.colorbar(label="Magnitude (dB)")
plt.ylim(0, 5)
plt.show()

# Plot for 100 ms window
plt.figure(figsize=(10, 6))
plt.pcolormesh(t_100ms, f_100ms_kHz[:f_index_limit_100ms], magnitude_100ms_dB[:f_index_limit_100ms, :], shading='gouraud', cmap='inferno')
plt.title("STFT Magnitude (100 ms)")
plt.ylabel("Frequency (kHz)")
plt.xlabel("Time (s)")
plt.colorbar(label="Magnitude (dB)")
plt.ylim(0, 5)
plt.show()
```

5. Describe some of the advantages and nackdelar of using short or long time windows for time-frequency analysis.

> We can see that the longer time window blurs in the time direction and the shorter time window blurs in the frequency direction. The longer windows will thus give more resolution for frequency. The shorter windows will give more resolution for timing.


For this case, we can see clearly that during 4 to 6 seconds, 25ms window lose much details in frequency direction, and during 6 to 8 seconds, 100ms window lose much details in time direction. So short time windows provide better time resolution, but poor frequency resolution, vice versa. We need to choose proper window to trade off time and frequency resolution while application.

6. For the first 10 seconds of your audio file (or shorter if it is not so long), use the [librosa package](https://github.com/librosa) to compute its Mel spectrogram using Hann windows of duration 25 ms with a window hopsize of 10 ms. Use 128 Mel bands and an FFT size of 8192 samples. Display the dB magnitude with reference to the max power observed, and limit your y-axis between 0 and 5 kHz. Use the same colormap as you used above. See https://github.com/librosa/librosa/blob/main/examples/LibROSA%20demo.ipynb for help. 

```python
import librosa
import librosa.display

# add your code below
# Load the audio file
audio_file = "fart.wav"
y, sample_rate = librosa.load(audio_file, sr=None)  # Load the file

# Parameters for the Mel spectrogram
fft_size = 8192                # FFT size
hop_size = int(0.01 * sample_rate)   # 10 ms hop size
n_mels = 128                   # Number of Mel bands
window_length = int(0.025 * sample_rate)  # 25 ms window

# Compute the Mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(
    y=y,
    sr=sample_rate,
    n_fft=fft_size,
    hop_length=hop_size,
    win_length=window_length,
    window='hann',
    n_mels=n_mels
)

# Convert the Mel spectrogram to dB, referenced to the max power observed
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

# Plot the Mel spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(
    mel_spectrogram_db,
    sr=sample_rate,
    hop_length=hop_size,
    x_axis='time',
    y_axis='mel',
    fmax=5000,  # Limit frequency axis to 5 kHz
    cmap='plasma'  # Use the 'plasma' colormap to match the earlier example
)

plt.colorbar(format='%+2.0f dB')
plt.title("Mel Spectrogram (25 ms Hann window, 128 Mel bands)")
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time (s)")
plt.ylim(0, 5000)  # Limit y-axis to 0–5 kHz
plt.show()
```
